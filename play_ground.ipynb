```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'train_images' contains the images
train_images = blood_data['train_images'][:6]

# Generate random noise
noise = np.random.normal(0, 0.02, train_images.shape)   # Mean 0, standard deviation 0.1

# Add noise to the images
train_images_noisy = train_images + noise

# Clip the values to be between 0 and 1
# train_images_noisy = np.clip(train_images_noisy, 0, 1)

# Plot the noisy images- show only the 80 imagesq
fig, axes = plt.subplots(10, 8,figsize=(10, 10))
axes = axes.ravel()

for i in range(80):
    axes[i].imshow(train_images_noisy[i], cmap='gray')
    axes[i].axis('off')

plt.tight_layout()
plt.show()

```
